```python
cat_cols=df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)     
print("Numerical Variables:")
print(num_cols)
```

```python
import seaborn as sns
for col in num_cols:
    print(col)
    print('Skew :', round(df[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    df[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Automatically determine number of rows and columns for subplot grid
n_cols = 2
n_rows = (len(cat_cols) + 1) // n_cols

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 6))
fig.suptitle('Bar Plot for All Categorical Variables in the Dataset', fontsize=20)
axes = axes.flatten()

# Plot each categorical column dynamically
for i, col in enumerate(cat_cols):
    order = df[col].value_counts().index[:20]  # Limit to top 20 categories for visibility
    sns.countplot(ax=axes[i], x=col, data=df, color='blue', order=order)
    axes[i].set_title(f'{col} Distribution')
    axes[i].tick_params(axis='x', rotation=45 if df[col].nunique() > 5 else 0)

# Hide any extra unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout to accommodate title and spacing
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import numpy as np

def log_transform(data, cols):
    for colname in cols:
        # Avoid negative or zero values for log
        if (data[colname] <= 0).any():
            # Shift values if necessary (e.g. min = 0 -> add 1)
            min_val = data[colname].min()
            shift = 1 - min_val if min_val <= 0 else 0
            data[colname + '_log'] = np.log(data[colname] + shift)
            print(f"[Shifted] Log transformed '{colname}' with shift = {shift}")
        else:
            data[colname + '_log'] = np.log(data[colname])
            print(f"[Log] Transformed '{colname}'")

    print("\nâœ… Transformation Complete.\n")
    data.info()
```

```python
# Optional: Auto-log transform skewed numeric features
from scipy.stats import skew

skewed = [col for col in num_cols if abs(df[col].skew()) > 1 and df[col].min() > 0]
log_transform(df, skewed)
```
