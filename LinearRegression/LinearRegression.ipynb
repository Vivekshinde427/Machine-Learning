```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes

df=load_diabetes()
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names
dataset.head()
```

```python
##Independant and Dependant

X=dataset
y=df.target
y
```

```python
## train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
## Standardizing the dataset

from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

##Cross Validation 

from sklearn.model_selection import cross_val_score

regression =LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
mse
```

```python
np.mean(mse)
```

```python
## Prediction 

reg_pred=regression.predict(X_test)
reg_pred
```

```python
import seaborn as sns
sns.distplot(reg_pred-y_test,kde=True)
```

```python
from sklearn.metrics import r2_score
score=r2_score(reg_pred,y_test)
score
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([3, 4, 2, 5, 6])

# Create model
model = LinearRegression()

# Train model
model.fit(x, y)

# Predict for input from user
x_input = float(input("Enter a value of X to predict Y: "))
x_input_array = np.array([[x_input]])
y_output = model.predict(x_input_array)

print(f"Predicted value of Y for X = {x_input} is: {y_output[0]:.2f}")

# Plot
y_pred = model.predict(x)
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted Line')
plt.scatter(x_input, y_output, color='green', label='Your Prediction', s=100)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Simple Linear Regression")
plt.show()

# Print coefficients
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```
